- from dataset we can conclude that 'day','month','year','Temperatue','RH','Ws' needed to be change to integer
- 'Rain','FFMC','DMC','DC','ISI','BUI','FWI' are needed to be changed into float datatype

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        if i in ['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']:
            df[i] = df[i].astype(float)
        elif i in ['day', 'month', 'year', 'Temperature', 'RH', 'Ws']:
            df[i] = df[i].astype(int)
df.info()
```
